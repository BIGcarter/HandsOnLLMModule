```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.special import erf
```

```python

# 定义激活函数
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def tanh(x):
    return np.tanh(x)

def relu(x):
    return np.maximum(0, x)

def leaky_relu(x, alpha=0.01):
    return np.where(x > 0, x, alpha * x)

def elu(x, alpha=1.0):
    return np.where(x > 0, x, alpha * (np.exp(x) - 1))

def softmax(x):
    exp_x = np.exp(x - np.max(x))
    return exp_x / np.sum(exp_x)

def swish(x, beta=1.0):
    return x * sigmoid(beta * x)

def gelu(x):
    return 0.5 * x * (1 + erf(x / np.sqrt(2)))

def silu(x):
    return x * sigmoid(x)

def softplus(x):
    return np.log(1 + np.exp(x))

def mish(x):
    return x * np.tanh(softplus(x))

def hard_sigmoid(x):
    return np.where(x <= -3, 0, np.where(x >= 3, 1, (x + 3) / 6))

def hard_tanh(x):
    return np.where(x <= -1, -1, np.where(x >= 1, 1, x))

# 生成输入数据
x = np.linspace(-10, 10, 1000)

# 绘制图像
plt.figure(figsize=(16, 12))

functions = [
    ('Sigmoid', sigmoid(x)),
    ('Tanh', tanh(x)),
    ('ReLU', relu(x)),
    ('Leaky ReLU', leaky_relu(x)),
    ('ELU', elu(x)),
    ('Softmax', softmax(x)),
    ('Swish', swish(x)),
    ('GELU', gelu(x)),
    ('SiLU', silu(x)),
    ('Softplus', softplus(x)),
    ('Mish', mish(x)),
    ('Hard Sigmoid', hard_sigmoid(x)),
    ('Hard Tanh', hard_tanh(x))
]

for i, (name, y) in enumerate(functions):
    plt.subplot(4, 4, i + 1)
    plt.plot(x, y, label=name)
    plt.title(name)
    plt.legend()

plt.tight_layout()
plt.show()
```
